## 1. Exercises for Lists

**1. Creating a List:**

Instructions: Create a list named my_fruits containing the following fruits: apple, banana, orange, and grape.

Hints: Use square brackets [] to create a list and separate items with commas.

**2. Accessing Elements:**

Instructions: Print the second fruit in the my_fruits list.

Hints: Use indexing to access elements, starting from 0.

**3. Adding Elements:**

Instructions: Add the fruit "mango" to the end of the my_fruits list.

Hints: Use the append() method to add elements to a list.

**4. Changing Elements:**

Instructions: Replace the third fruit in the my_fruits list with "strawberry".

Hints: Use indexing to access and modify specific elements.

**5. Removing Elements:**

Instructions: Remove the first fruit from the my_fruits list.

Hints: Use the pop() method to remove elements by index.

**6. Finding Length:**

Instructions: Print the number of fruits in the my_fruits list.

Hints: Use the len() function to find the length of a list.

**7. Slicing Lists:**

Instructions: Create a new list called first_two_fruits containing only the first two fruits from the my_fruits list.

Hints: Use slicing to extract a portion of a list.

**8. Sorting a List:**

Instructions: Alphabetically sort the my_fruits list in ascending order.

Hints: Use the sort() method to sort a list.

**9. Reversing a List:**

Instructions: Reverse the order of the my_fruits list.

Hints: Use the reverse() method to reverse a list's elements.

**10. Finding a Value:**

Instructions: Check if the fruit "banana" exists in the my_fruits list.

Hints: Use the in operator to check for membership in a list.

**11. Nested Lists:**

Instructions: Create a list of grocery items, where each item is itself a list containing the product name and quantity. For example, groceries = [["apples", 3], ["milk", 2], ["bread", 1]].

Hints: Lists can contain other lists as elements.

**12. Iterating Through Lists:**

Instructions: Print each fruit in the my_fruits list, one fruit per line.

Hints: Use a for loop to iterate through the elements of a list.

## 2. Exercises for Dictionaries

**1. Creating a Dictionary:**

Instructions: Create a dictionary named student_info with the following key-value pairs: name = "Alice", age = 25, city = "New York".\
Hints: Use curly braces {} to create a dictionary and colons : to separate keys and values.

**2. Accessing Values:**

Instructions: Print the value associated with the key "age" in the student_info dictionary.\
Hints: Use square brackets [] to access values by their keys.

**3. Adding Key-Value Pairs:**

Instructions: Add a new key-value pair to the student_info dictionary: major = "Computer Science".\
Hints: Assign a value to a new key to add it to a dictionary.

**4. Modifying Values:**

Instructions: Change the value of the "city" key in the student_info dictionary to "London".\
Hints: Assign a new value to an existing key to modify it.

**5. Removing Key-Value Pairs:**

Instructions: Remove the key-value pair with the key "age" from the student_info dictionary.\
Hints: Use the pop() method to remove a key-value pair by its key

**6. Iterating Through Key-Value Pairs:**

Instructions: Loop through each key-value pair in the student_info dictionary and print the key followed by its value.\
Hints: Use a for loop to iterate through the items of a dictionary.

**7. Checking for Keys and Values:**

Instructions: Check if the key "major" exists in the student_info dictionary.\
Hints: Use the in operator to check for key existence.

**8. Merging Dictionaries:**

Instructions: Create a new dictionary named course_info with the following key-value pairs: course_name = "Python Programming", professor = "Dr. Smith", semester = "Spring 2024". Merge this dictionary with the student_info dictionary.\
Hints: Use the update() method to merge one dictionary into another.

**9. Nested Dictionaries:**

Instructions: Create a dictionary named family with the key "members" and value as another dictionary containing key-value pairs for each family member (name, age, relationship).\
Hints: Dictionaries can contain other dictionaries as values.

**10. Dictionary Comprehensions:**

Instructions: Create a new dictionary named squares that maps each number from 1 to 10 to its square (e.g., {1: 1, 2: 4, 3: 9}).\
Hints: Use dictionary comprehensions to create dictionaries with concise syntax.

**11. String Manipulation:**

Instructions: Given the string "Hello World!", perform the following operations:
Convert all letters to lowercase.
Replace "World" with "Python".
Extract the substring from "Hello" to (but not including) the space.\
Hints: Use string methods like lower(), replace(), and split() to manipulate strings.

**12. Regular Expressions:**

Instructions: Use a regular expression to match all occurrences of the word "apple" in the following string: "The apple is red, the apple is sweet, but the apple fell from the tree."\
Hints: Learn basic regular expression syntax to find patterns in strings.

## 2. Exercises for Strings

**1. Creating Strings:**

Instructions: Create a string variable named my_name and assign your name to it.\
Hints: Use single or double quotes to enclose strings.

**2. Accessing Characters:**

Instructions: Print the first letter of your name from the my_name string.\
Hints: Use indexing to access individual characters, starting from 0.

**3. String Length:**

Instructions: Print the number of characters in your name using the len() function.

**4. Slicing Strings:**

Instructions: Extract your first name from the my_name string, assuming it's in the format "first_name last_name".\
Hints: Use slicing to extract a portion of a string.

**5. Concatenation:**

Instructions: Create a greeting message by combining "Hello, " with your name.\
Hints: Use the + operator to concatenate strings.

**6. String Formatting:**

Instructions: Print the following sentence using string formatting: "My favorite number is {number}". Replace {number} with your actual favorite number.\
Hints: Use f-strings or the format() method to insert variables into strings.

**7. Uppercase and Lowercase:**

Instructions: Convert your name to uppercase and print it. Then convert it back to lowercase and print it again.\
Hints: Use the upper() and lower() methods to change the case of strings.

**8. Finding Substrings:**

Instructions: Check if your name contains the letter "a".\
Hints: Use the in operator to check for substrings.

**9. Replacing Substrings:**

Instructions: Replace all occurrences of the letter "e" in your name with the letter "x".\
Hints: Use the replace() method to substitute parts of a string.

**10. Splitting Strings:**

Instructions: Split the sentence "This is a sentence with multiple words" into a list of words.\
Hints: Use the split() method to divide a string based on a delimiter.

**11. String Reversal:**

Instructions: Reverse the order of characters in your name and print the reversed string.\
Hints: Explore different techniques like slicing or string manipulation to reverse a string.

**12. Palindrome Check:**

Instructions: Write a function that takes a string as input and returns True if it's a palindrome (reads the same backward as forward), False otherwise.\
Hints: Consider using slicing and comparison to check for palindromes.
